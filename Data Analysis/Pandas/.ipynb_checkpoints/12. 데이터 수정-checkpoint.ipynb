# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## Column 수정

```python
df['학교'].replace({'북산고':'상북고', '능남고':'비비고'})
```

```python
df
```

```python
df['학교'].replace({'북산고':'비비고'}, inplace=True)
```

```python
df
```

```python
df['SW특기'].str.lower()
```

```python
df['SW특기'] = df['SW특기'].str.lower() # 원하는 값을 변수로 넣어주면 바로 적용됨
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교'] + '등학교' # 학교데이터 + 등학교
df
```

## Column 추가

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회'] # 새 Column 추가
df
```

```python
df['결과'] = 'Fail' # 결과 Column 을 추가하고 전체 데이터는 Fail 로 초기화
df
```

```python
df.loc[df['총합'] > 400, '결과'] = 'Pass' # 총합이 400 보다 큰 데이터를 Pass 로 업데이트
df
```

## Column 삭제

```python
df.drop(columns=['총합']) # 총합 Column 을 삭제
```

```python
df.drop(columns=['국어', '영어', '수학']) # 국영수 삭제 (inplace=True 하면 반영됨)
```

```python
df
```

## Row 삭제

```python
df.drop(index='4번') # 4번 학생 데이터 row 삭제
```

```python
filt = df['수학'] < 80 # 수학 점수가 80점 미만인 학생 필터링
df[filt]
```

```python
df[filt].index
```

```python
df.drop(index=df[filt].index) # filt 에 해당하는 index 위치의 row 필터링
```

## Row 추가

```python
df.loc['9번'] = ['이정환', '해남고등학교', 184, 90, 90, 90, 90, 90, 'Kotlin', 450, 'Pass'] # 새로운 Row 추가
df
```

## Cell 수정

```python
df.loc['4번', 'SW특기'] = 'Python' # 서태웅 특기를 파이썬으로
df
```

```python
df.loc['5번', ['학교', 'SW특기']] = ['능남고등학교', 'C'] # 강백호 학교, 특기 변경
df
```

## Column 순서 변경

```python
cols = list(df.columns)
cols
```

```python
df = df[[cols[-1]] + cols[0:-1]] # 맨 뒤에 있는 결과 Column 을 앞으로 가져오고, 나머지 Column 들과 합쳐서 순서 변경
df
```

```python
df = df[['결과', '이름', '학교']]
df
```
